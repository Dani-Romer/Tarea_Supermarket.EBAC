# 1.- Generar estadística descriptiva básica en las columnas unit_price y quantity: Cálculo de la media, mediana, moda

```python
unit_price_np = df[['Unit price']].to_numpy()
quantity_np = df[['Quantity']].to_numpy()
```

```python
media_price = unit_price_np.mean()
media_quantity = quantity_np.mean()

format_m_price = '{:.2f}'.format(media_price)
format_m_quantity = '{:.2f}'.format(media_quantity)

print ('La media de unit price es = ', format_m_price)
print ('La media de quantity es = ', format_m_quantity)       # Media de unit_price y quantity
```

```python
vals1, counts1 = np.unique(unit_price_np, return_counts = True)
index1 = np.argmax(counts1)
moda_price = vals1[index1]

print ('La moda del unit price es = ', moda_price)                 # moda de unit price
```

```python
vals2, counts2 = np.unique(quantity_np, return_counts = True)
index2 = np.argmax(counts2)
moda_quantity = vals1[index2]

print ('La moda del quantity es = ', moda_quantity)                 # moda de quantity
```

```python
mediana_price = np.median (unit_price_np)
mediana_quantity = np.median (quantity_np)

print (f'La mediana de unit price es = {mediana_price:.2f}')
print (f'La mediana de quantity es = {mediana_quantity:.2f}')      # Mediana de unit_price y quantity
```

# Obtener el ticket promedio:
#1) por ciudad,
#2) por product line y ciudad,
#3) por género,
#4) por género y product line

```python
city_np = df[['City']].to_numpy()
product_np = df[['Product line']].to_numpy()
gender_np = df[['Gender']].to_numpy()
total_np = df[['Total']].to_numpy()
```

```python
ciudades_u = np.unique(city_np)

for ciudad in ciudades_u:
    ventas_ciudad = total_np[city_np == ciudad]
    p_ventas_ciudad = np.mean(ventas_ciudad)
    print(f'La venta promedio en {ciudad} es {p_ventas_ciudad:.2f}')       # Venta total promedio por ciudad
```

```python
city_u = np.unique(city_np)
product_u = np.unique(product_np)

for ciudad in city_u:
    for tipo in product_u:
        ventas = total_np[(city_np == ciudad) & (product_np == tipo)]
        if ventas.size > 0:
            promedio_c_p = np.mean(ventas)
            print(f'Venta promedio en {ciudad} para {tipo}: {promedio_c_p:.2f}')    # Venta total promedio por ciudad y por Product line
```

```python
gender_u = np.unique(gender_np)

for gender in gender_u:
    ventas_gender = total_np[gender_np == gender]
    p_ventas_gender = np.mean(ventas_gender)
    print(f'La venta promedio por genero {gender} es {p_ventas_gender:.2f}')   # promedio total de venta por genero
```

```python
gender_u = np.unique(gender_np)
product_u = np.unique(product_np)

for gender in gender_u:
    for tipo in product_u:
        ventas = total_np[(gender_np == gender) & (product_np == tipo)]
        if ventas.size > 0:
            promedio_g_p = np.mean(ventas)
            print(f'Venta promedio por genero {gender} para {tipo}: {promedio_g_p:.2f}')  # promedio total de venta por genero y por Product line
```

# Usar el método rank para generar top 5 de:
#1) ventas por ciudad,
#2) ventas por member,
#3) ventas por payment.  Obtener además el % de aporte de cada categoría. 

```python
df_v_ciudad = df.sort_values(by=['City', 'Total'], ascending=[True, False])
```

```python
top5_v_ciudad = df_v_ciudad.groupby('City').head(5)
```

```python
top5_v_ciudad    # Top 5 de ventas por ciudad
```

```python
total_ventas = df['Total'].sum()
ventas_por_ciudad = df.groupby('City')['Total'].sum()

porcentaje_v_ciudad = (ventas_por_ciudad / total_ventas) * 100

print('El Porcentaje de ventas por ciudad es = ', porcentaje_v_ciudad)            # porcentaje de ventas por ciudad
```

```python
df_c_type = df.sort_values(by=['Customer type', 'Total'], ascending=[True, False])
top5_c_type = df_c_type.groupby('Customer type').head(5)

top5_c_type                                # top 5 de ventas mas grandes por customer type
```

```python
total_ventas = df['Total'].sum()
ventas_c_type = df.groupby('Customer type')['Total'].sum()

porcentaje_c_type = (ventas_c_type/ total_ventas) * 100

print('El Porcentaje de ventas por customer type = ', porcentaje_c_type)            # porcentaje de ventas por ciudad
```

```python
df_payment = df.sort_values(by=['Payment', 'Total'], ascending=[True, False])
top5_payment = df_payment.groupby('Payment').head(5)

top5_payment                                         #top 5 de ventas mas grandes por payment
```

```python
total_ventas = df['Total'].sum()
ventas_payment = df.groupby('Payment')['Total'].sum()

porcentaje_payment = (ventas_payment/ total_ventas) * 100

print('El Porcentaje de ventas por Payment = ', porcentaje_payment)       # porcentaje de ventas por payment
```

# Cálculo de una matriz de correlación entre la hora (sin minutos) y el total, y otra que correlacione el unit_price con el rating de la transacción, para validar si los productos más caros son los que dejan más margen

```python
df.info ()
```

```python
from datetime import datetime

df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
```

```python
df['Time'] = df['Time'].dt.hour
```

```python
df
```

```python
total_np = df[['Total']].to_numpy()
time_np = df[['Time']].to_numpy()
```

```python
total_np = np.concatenate (total_np, axis=0)
time_np = np.concatenate (time_np, axis=0)
```

```python
cor_time_total = np.corrcoef(total_np, time_np)
```

```python
cor_time_total                               # correlacion ente time y el total
```

```python
unit_p_np = df[['Unit price']].to_numpy()
rating_np = df[['Rating']].to_numpy()
```

```python
unit_p_np = np.concatenate (unit_p_np, axis=0)
rating_np = np.concatenate (rating_np, axis=0)
```

```python
cor_unit_rating = np.corrcoef(unit_p_np, rating_np)
```
